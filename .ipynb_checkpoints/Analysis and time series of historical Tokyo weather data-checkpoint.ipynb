```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

from datetime import datetime
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('tokyo_weather_cleaned.csv')
df_gas = pd.read_csv('japan_greenhouse_gas_levels.csv')
```

```python
df.head()
```

```python
df.set_index('Date')
df.index
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.Date
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
# daily temperature range
df['Temp variation'] = df.apply(lambda x: x['Max temp'] - x['Min temp'], axis=1)
df['Temp variation']
```

```python
df_gas.head()
```

```python
df_gas.rename(columns={df_gas.columns[0]: 'Date'}, inplace=True)
df_gas['Date'] = pd.to_datetime(df['Date'])
df_gas['Year'] = pd.DatetimeIndex(df_gas['Date']).year
df_gas['Month'] = pd.DatetimeIndex(df_gas['Date']).month
df_gas.tail()
```

# Historical temperature analysis (1981-2021)

```python
df.plot(x='Date',y='Max temp',style='.', figsize=(15,5), color=color_pal[0], ylabel='Temperature °C')
plt.title('Max Temperature')
plt.legend('', frameon=False)
```

```python
df.plot(x='Date',y='Min temp',style='.', figsize=(15,5), color=color_pal[1], ylabel='Temperature °C')
plt.title('Min Temperature')
plt.legend('', frameon=False)
```

```python
df.plot(x='Date',y='Ave temp',style='.', figsize=(15,5), color=color_pal[2], ylabel='Temperature °C')
plt.title('Ave Temperature')
plt.legend('', frameon=False)
```

```python
max_T_yr = df.groupby('Year')[['Max temp']].max()
max_T_yr.plot(style='-', figsize=(15,5), color=color_pal[0], ylabel='Temperature °C', rot=45)
plt.xticks(np.arange(1981,2021,1))
plt.title('Max Temperature by Year')
plt.legend('', frameon=False)
```

```python
ave_T_yr = df.groupby('Year')[['Min temp']].max()
ave_T_yr.plot(style='-', figsize=(15,5), color=color_pal[1], ylabel='Temperature °C', rot=45)
plt.xticks(np.arange(1981,2021,1))
plt.title('Highest Minimum Temperature by Year')
plt.legend('', frameon=False)
```

```python
ave_T_yr = df.groupby('Year')[['Ave temp']].max()
ave_T_yr.plot(style='-', figsize=(15,5), color=color_pal[2], ylabel='Temperature °C', rot=45)
plt.xticks(np.arange(1981,2021,1))
plt.title('Highest Average Temperature by Year')
plt.legend('', frameon=False)
```

```python
range_T_yr = df.groupby('Year')[['Temp variation']].max()
range_T_yr.plot(style='-', figsize=(15,5), color=color_pal[2], ylabel='Temperature °C', rot=45)
plt.xticks(np.arange(1981,2021,1))
plt.title('Highest Day/Night Temperature Range by Year')
plt.legend('', frameon=False)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=df,x='Year', y='Temp variation')
plt.ylabel('Day/Night T variation °C')
plt.xticks(rotation=45)
plt.show()
```

```python
df_gas.plot(x='Date', y='CO2_ppm', ylabel='CO2 ppm',color=color_pal[3])
plt.title('Carbon Dioxide Emissions')
plt.legend('', frameon=False)
```

# Max Temperature Train/Test Split

```python
train = df.loc[df.Date < '2018-12-31']
test = df.loc[df.Date > '2019-01-01']

fig, ax= plt.subplots(figsize=(15,5))
train.plot(ax=ax, label='Training Set (1981-2018)', x='Date', y='Max temp')
test.plot(ax=ax, label='Test Set (2019-2021)', x='Date', y='Max temp')
plt.show()
```

# Ave Temperature Train/Test Split

```python
train = df.loc[df.Date < '2018-12-31']
test = df.loc[df.Date > '2019-01-01']

fig, ax= plt.subplots(figsize=(15,5))
train.plot(ax=ax, label='Training Set (1981-2018)', x='Date', y='Ave temp')
test.plot(ax=ax, label='Test Set (2019-2021)', x='Date', y='Ave temp')
plt.show()
```
