```python
#making a consistent date format of yyyy-mm-dd
    #replace / with -
    #add a 0 in front of month or day if they are a single digit

def dates(date):
    splits = date.split('/')
    
    if len(splits[1]) == 1:
        splits[1] = '0' + splits[1]
        
    if len(splits[2]) == 1:
        splits[2] = '0' + splits[2]
        
    x = splits[0] + '-' + splits[1] + '-' + splits[2]
    
    return x
```

# Regression model training with cross-validation and time lag features - Max Temperature data

```python
max_df = df.drop(columns=['Ave temp', 'Min temp'])
```

## Lag features

```python
def add_lag(df, target):
    #1, 2, 3, 5, 10, 15 year lag
    target_map = df[target].to_dict()
    df[f'lag_1_{target}'] = (df.index - pd.Timedelta(364, 'd')).map(target_map) #1 year
    df[f'lag_2_{target}'] = (df.index - pd.Timedelta(728, 'd')).map(target_map) #2 years
    df[f'lag_3_{target}'] = (df.index - pd.Timedelta(1092,'d')).map(target_map) #3 years
    df[f'lag_4_{target}'] = (df.index - pd.Timedelta(1820,'d')).map(target_map) #5 years
    df[f'lag_5_{target}'] = (df.index - pd.Timedelta(3640,'d')).map(target_map) #10 years
    df[f'lag_6_{target}'] = (df.index - pd.Timedelta(5460,'d')).map(target_map) #15 years
    
    return df
```

```python
max_df = add_lag(max_df, 'Max temp')
max_df.tail(10)
```

```python
#train 10 folds and score the model

tss = TimeSeriesSplit(n_splits=10, test_size = 365*1, gap=1)
max_df = max_df.sort_index()

preds = []
scores = []

for train_idx, val_idx in tss.split(max_df):
    train = max_df.iloc[train_idx]
    test = max_df.iloc[val_idx]
    

    FEATURES = ['Week','Month','Quarter','Year','lag_1_Max temp','lag_2_Max temp','lag_3_Max temp',
               'lag_4_Max temp','lag_5_Max temp','lag_6_Max temp']
    TARGET = 'Max temp'
    
    X_train = train[FEATURES]
    Y_train = train[TARGET]
    X_test = test[FEATURES]
    Y_test = test[TARGET]
    
    max_reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                          n_estimators=1000, 
                          early_stopping_rounds=50,
                          objective='reg:squarederror',
                          max_depth=5,
                          learning_rate=0.01,
                          colsample_bytree=0.8,
                          subsample=0.7)
    
    max_reg.fit(X_train, Y_train,
           eval_set=[(X_train, Y_train),(X_test, Y_test)],
           verbose=100)
    
    Y_pred = max_reg.predict(X_test)
    preds.append(Y_pred)
    #score the model - mean squared error gives penalties to the model
    score = np.sqrt(mean_squared_error(Y_test, Y_pred))
    scores.append(score)

```

```python
mean_mse = np.mean(scores)
print(f'Mean MSE across folds: {mean_mse}')
print(f'MSE: {scores}')
```

```python
mean_rmse = pow(mean_mse,0.5)
mean_rmse
```

```python
test['Prediction_MAX'] = max_reg.predict(X_test)
test_max_df = max_df.merge(test['Prediction_MAX'], how='left', left_index=True, right_index=True)
test_max_df
```

# Forecasting max temperatures with 5 splits and lag features - predict 3 years into the future

```python
#training on all data
FEATURES = ['Week','Month','Quarter','Year','lag_1_Max temp','lag_2_Max temp','lag_3_Max temp',
               'lag_4_Max temp','lag_5_Max temp','lag_6_Max temp']
TARGET = 'Max temp'

X_all = max_df[FEATURES]
Y_all = max_df[TARGET]

max_all_reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                          n_estimators=1000, 
                          early_stopping_rounds=50,
                          objective='reg:squarederror',
                          max_depth=5,
                          learning_rate=0.01,
                          colsample_bytree=0.8,
                          subsample=0.7)
max_all_reg.fit(X_all, Y_all,
       eval_set=[(X_all, Y_all)],
       verbose=100)
```

```python
max_forecast = pd.date_range(start='2024-06-27', end='2027-06-27', freq='D')
max_forecast_df = pd.DataFrame(index=max_forecast)
max_forecast_df['isFuture'] = True
max_df['isFuture'] = False

df_and_forecast = pd.concat([max_df,max_forecast_df])
df_and_forecast = create_features(df_and_forecast)
df_and_forecast = add_lag(df_and_forecast, 'Max temp')
df_and_forecast
```

```python
max_future_features = df_and_forecast.query('isFuture').copy()
max_future_features['MAX Pred']=max_all_reg.predict(max_future_features[FEATURES])
max_future_features
```

```python
max_future_features['MAX Pred'].plot(x='Date',
                                    color=color_pal[1],
                                    ms=1,
                                    lw=1,
                                    title='3 Year Max Temp Forecast')
plt.ylabel('°C')
plt.show()
```

```python
#train 10 folds and score the model

tss = TimeSeriesSplit(n_splits=10, test_size = 365*1, gap=1)
df = ave_df.sort_index()

preds = []
scores = []
for train_idx, val_idx in tss.split(df):
    train = ave_df.iloc[train_idx]
    test = ave_df.iloc[val_idx]
    

    FEATURES = ['Week','Month','Quarter','Year','lag_1_Ave temp','lag_2_Ave temp','lag_3_Ave temp',
               'lag_4_Ave temp','lag_5_Ave temp','lag_6_Ave temp']
    TARGET = 'Ave temp'
    
    X_train = train[FEATURES]
    Y_train = train[TARGET]
    X_test = test[FEATURES]
    Y_test = test[TARGET]
    
    reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                          n_estimators=500, 
                          early_stopping_rounds=50,
                          objective='reg:squarederror',
                          max_depth=5,
                          learning_rate=0.01)
    reg.fit(X_train, Y_train,
           eval_set=[(X_train, Y_train),(X_test, Y_test)],
           verbose=100)
    
    Y_pred = reg.predict(X_test)
    preds.append(Y_pred)
    #score the model - mean squared error gives penalties to the model
    score = np.sqrt(mean_squared_error(Y_test, Y_pred))
    scores.append(score)
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
            index=reg.feature_names_in_,
            columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.legend('', frameon=False)
plt.show()
```

```python
print(f'Score across folds: {np.mean(scores)}')
print(f'Fold scores: {scores}')
```
